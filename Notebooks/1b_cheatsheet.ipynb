## Functions and Data Visualization

### Functions

In Python, a function is a reusable block of code that performs a specific task when called. Functions are used to organize code into smaller, manageable units, making your programs more readable and maintainable. Here's an explanation of how functions work in Python.

- Function Definition: To create a function, you use the ```def``` keyword followed by the function name and parentheses, which can include parameters. The function definition ends with a colon.

                    Example :
                         def greet(name):
                             print("Hello, " + name)
                         
- Function Parameters: Parameters are variables that you can pass to a function. They allow you to provide input to the function, which it can then use in its operations. In the example above, ```name``` is a parameter.

- Function Body: The function body is indented and contains the code that the function executes when called. You can include any valid Python code within the function body.

- Function Call: To use a function, you call it by its name followed by parentheses, and you can pass arguments that match the parameters defined in the function.

                   Example:
                          greet("Alice")
                          
- Retun Statement: Functions can return values using the return statement. The returned value can be assigned to a variable or used in expressions. The convention is to put a blank line before a function definition as well as after one.

                   Example:
                        def add(a, b):
                            return a + b
                        
                        result = add(3, 5)   # Result will be 8

### Conditionals

The ``if`` statement is a conditional statement in Python used to determine whether a block of code will be executed. If the program finds the condition defined in the if statement `True`, it will execute the code block inside the if statement. If the conditional expression results in a `False` value, the statement below will not be executed.

In python, if-statements will always have the following form:

``if <expr>:
    <statement>``

In the form shown above:
- ``<expr>`` is a Boolean expression, that should return either ``TRUE`` or ``FALSE``. This is the conditional that is assessed in order to continue with the lines below.
- ``<statement>`` is a valid Python statement, which must be indented.

Examples of conditional statements are:
- ``if x>y:`` ,
if variable x is larger than variable y
- ``if a==b:`` ,
if variable a is equal to variable b
- ``if v in [1,2,3,4,5]:`` ,
if variable v is in a list with the numbers 1-5
- ``if w != 6:`` ,
if variable w is not equal to the value 6

Note that the syntax for 'is equal to' is `==`in Python, and the syntax for 'is not equal to' is `!=`.

In addition to the `if` statement above, Python also allows you to define an `else` statement. This will have the following form:

``if <expr>:
      <statement 1>
  else:
      <statement 2>``

If the conditional expression `<expr>` will result in a `True` value, `<statement 1>` will be executed. If `<expr>` will be `False`, the programme will continue with `<statement 2>`. Adding an `else` statement to your code will therefore give you the opportunity to tell the programme what it has to do in the other cases (if they appear). We will show you practical examples in the exercises.

### For loops

A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string). With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc. This for loop can look as follows:

``for i in <list>:
    <statement>``

The new variable ``i`` will take every value in the ``<list>`` or other type of sequence. For each value, the ``<statement>`` below is executed. This makes it possible to execute the statement multiple times automatically, without the need for the same lines of code. With the example exercises below, you can practice using the for loop.

Examples of conditional statements are:
- ``for i in [0,1,2,3,4,5]:`` ,
where the newly created variable ``i`` will take the value ``0`` in the first loop, and execute the statement below. After that, ``i`` will be equal to ``1`` and execute the same statement, and so on.
- ``for name in ['Flora','Jacques','Ruud','Krishna','Siham','Lotte']`` ,
where the newly created variable ``name`` will take the value ``Flora`` in the first loop, and execute the statement below. After that, ``name`` will be equal to ``Jacques`` and execute the same statement, and so on.

### `range()` function

In Python, there is a predefined function that can help us create a range with ascending or descending integers. It is often used in combination with for loops, such that the iteration variable `i` will take values in this range.

The `range()` function requires a start and a stop value `range(<start>,<stop>)`. It will count from the start value to the stop value, where the stop value will not_ be included! As an example, the `range(0,5)` will give a range as: 0, 1, 2, 3, 4. For more explanation, see https://www.w3schools.com/python/ref_func_range.asp .

The start value is optional, so if you only use a stop value, the range will start at 0 by default. This means that `range(0,10)` and `range(10)` both give the same results. To show you how the range function works in combination with a for loop, see the examples below.


````python
for i in range(5):
    print(i)
````

````python
for j in range(3,9):
    print(j)
````

````python
#Without a start value, the range function will take 0 as default start value
print(range(5))
````

### Creating an `AmountNet` column
In some GL datasets, the amount is always positive (>=0), with a separate column indicating whether it is a debit or credit amount. This can be inconvenient for your analysis, because you can't summarize those amounts without considering the debit/credit column.

To make your analysis easier, you can create a new column `AmountNet` that contains the net amount, i.e., the amount as a positive number for debits and a negative number for credits.

In Excel, you would fix this with a formula like this: `=IF(B2="Credit",C2*-1,C2)`. You can also easily fix this with Python, using the `loc` function we discussed in the previous lecture:

```
# Set AmountNet to 1x the Amount for Debit transactions
gl_data.loc[gl_data['Debit/Credit'] == 'Debit', 'AmountNet'] = gl_data['Amount']
# Set AmountNet to 1x the Amount for Credit transactions
gl_data.loc[gl_data['Debit/Credit'] != 'Debit', 'AmountNet'] = gl_data['Amount'] * -1
```


### The GL/Journal matrix using pivot_table
The **GL/Journal matrix** is a table that shows the Amount per GL account and Journal. This table can be used to quickly identify patterns and anomalies in the data.

We can create it easily using Python and Pandas:
```
# Pivot the data to create a GL/Journal matrix
gl_df.pivot_table(index='Account', columns='Journal', values='AmountNet', aggfunc='sum', margins=True).fillna(0)
```

A few remarks:
* `index='Account'` means that the GL account will be the rows of the matrix.
* `columns='Journal'` means that the Journal will be the columns of the matrix.
* `values='AmountNet'` means that the AmountNet column will be used to fill the matrix. This is the value that will be summarized.
* `aggfunc='sum'` means that the values will be summarized. You can also use other functions, like `mean`, `count`, etc. Do not forget this option, by default panda assumes `mean`...
* `margins=True` requests Pandas to add 'totals' to the matrix.


You can also use two or more columns for the Index or Columns values. For example, to create a GL/Journal matrix with the Account and AccountType as the Index, and the Period as the Columns, you can use the following code:
```
# Pivot the data to create a GL/Journal matrix with Account and AccountType as the index
gl_df.pivot_table(index=['Account', 'AccountType'], columns='Period', values='AmountNet', aggfunc='sum').fillna(0)
```

Notice how we changed `index='Account'` (`'Account'` is a single string value) to `index=['Account', 'AccountType']` (here `['Account', 'AccountType']` is a list of two strings!). This is how you can use multiple columns for the Index or Columns values.
