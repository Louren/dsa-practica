### 3.1 Understanding the data

Q3.1.1 We will continue analyzing the GL transactions in `gl_data`, the dataset with *all* journal transactions from DigiJazz. We will focus on the `Amount` column. Try to understand the data by using the `describe()` function. What do you see? Hint: think about debit and credit

````python
# A3.1.1
# We see that all the values in the 'Amount' column are positive. That is 'strange' for a general ledger,
# as we would expect to see both positive (debit) and negative (credit) values.
# However, we often see this. In a separate column, the 'Debit/Credit' column, we would expect to see the type of transaction.
gl_data['Amount'].describe()
````

Did you observe that all amounts are positive? This is because the dataset is a general ledger, where all transactions are recorded as positive amounts. The debit and credit are indicated by the `Debit` and `Credit` columns.

For our analysis, a common starting point is to calculate the *net amount per transaction*. For all Credit transactions, the `Amount` is multiplied by `-1`.

````python
# Set AmountNet to 1x the Amount for Debit transactions
gl_data.loc[gl_data['Debit/Credit'] == 'Debit', 'AmountNet'] = gl_data['Amount']
# Set AmountNet to 1x the Amount for Credit transactions
gl_data.loc[gl_data['Debit/Credit'] != 'Debit', 'AmountNet'] = gl_data['Amount'] * -1

# Check the results
gl_data['AmountNet'].describe()
````

Q3.1.2. The GL/Journal matrix is a table that shows the Amount per GL account and Journal. This table can be used to quickly identify patterns and anomalies in the data. Create a GL/Journal matrix using Python. Refer to Cheatsheet 1b if you like more information on how to do this.

````python
# The GL/Journal matrix is a table that shows the Amount per GL account and Journal.
# This table can be used to quickly identify patterns and anomalies in the data.
# We can create it easily using Python and Pandas:

# Pivot the data to create a GL/Journal matrix
gl_journal_matrix = gl_data.pivot_table(index='AccountDesc', columns='JournalDesc', values='AmountNet', aggfunc='sum').fillna(0)

# Show the GL/Journal matrix
gl_journal_matrix
````

Q3.1.3. Another common analysis is to calculate the *amount per GL account*, per *period*. This gives insight into the revenue and costs per period. Using the same `pivot_table` function, create a table that shows the `Amount` per `GL Account` and `Period`.

````python
# A3.1.3
pivot = gl_data.pivot_table(index=['AccountDesc'], columns='Period', values='AmountNet', aggfunc='sum').fillna(0).round(0)
pivot
````

# 3.3 Margin Analysis

As part of our risk analysis, we want to understand the sales margins. We can calculate the margin by subtracting the costs from the revenue. We can then compare the observed margins to margins that management has authorized.

We will introduce a few new techniques, that you don't all need to know by heart. We will help you with example code, that you should finalize.

````python
sal = gl_data[gl_data['JournalType'] == 'Revenue']
sal
````

````python
# We use the pyplot and seaborn libraries to visualize the data, refer to Cheatsheet Week 1a - Libraries and Exploring a dataset
import matplotlib.pyplot as plt
import seaborn as sns


# Analyzing transaction values for outliers
# Define thresholds based on typical transaction ranges
threshold_lower = 30 # Assumed min 30% margin based on knowledge from DigiJazz
threshold_upper = 75 # Assumed max 75% margin based on knowledge from DigiJazz

# Group transactions to see the distribution
grouped = sal.groupby(['Product ID'])['Amount'].describe()
print("\nTransaction summary by Product ID:")
print(grouped)

# Plot histogram for transaction values
plt.figure(figsize=(8, 6))
sns.histplot(sal['Amount'], bins=20, kde=True)
plt.axvline(threshold_lower, color='red', linestyle='--', label='Lower Threshold')
plt.axvline(threshold_upper, color='blue', linestyle='--', label='Upper Threshold')
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Amount')
plt.legend()
plt.show()
````
