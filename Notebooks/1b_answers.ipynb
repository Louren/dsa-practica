Q3.1.1 We will continue analyzing the GL transactions in `gl_data`, the dataset with *all* journal transactions from DigiJazz. We will focus on the `Amount` column. Try to understand the data by using the `describe()` function. What do you see? Hint: think about debit and credit and +/- signs.

````python
# A3.1.1
# We see that all the values in the 'Amount' column are positive. That is 'strange' for a general ledger,
# as we would expect to see both positive (debit) and negative (credit) values.
# However, we often see this. In a separate column, the 'Debit/Credit' column, we would expect to see the type of transaction.
gl_data['Amount'].describe()
````

Did you observe that all amounts are positive? This is because the dataset is a general ledger, where all transactions are recorded as positive amounts. The debit and credit are indicated by the `Debit` and `Credit` columns.

For our analysis, a common starting point is to calculate the *net amount per transaction*. For all Credit transactions, the `Amount` is multiplied by `-1`.

````python
# Set AmountNet to 1x the Amount for Debit transactions
gl_data.loc[gl_data['Debit/Credit'] == 'Debit', 'AmountNet'] = gl_data['Amount']
# Set AmountNet to 1x the Amount for Credit transactions
gl_data.loc[gl_data['Debit/Credit'] != 'Debit', 'AmountNet'] = gl_data['Amount'] * -1

# Check the results
gl_data['AmountNet'].describe()
````

Refer to Cheatsheet 1b if you like more information on how to do this.

````python
# A3.1.2
# The GL/Journal matrix is a table that shows the Amount per GL account and Journal.
# This table can be used to quickly identify patterns and anomalies in the data.
# We can create it easily using Python and Pandas:

# Pivot the data to create a GL/Journal matrix
gl_journal_matrix = gl_data.pivot_table(index='Account', columns='JournalDesc', values='AmountNet', aggfunc='sum').fillna(0)

# Here we use 'fillna' to replace 'na' values with 0, that looks better in the output
gl_journal_matrix = gl_journal_matrix.fillna(0)

# Show the GL/Journal matrix
gl_journal_matrix
````

### 3.4 Margin Analysis

As part of our risk analysis, we want to understand the sales margins. We can calculate the margin by subtracting the costs from the revenue. We can then compare the observed margins to margins that management has authorized.

We will introduce a few new techniques, that you don't all need to know by heart. We will help you with example code, that you should finalize.

You can run all of this code without any changes. But try to understand! You can then use the results to answer the question at the end of the notebook.


````python
# Calculate margin per order (using the 'Source_id' column)
salmargin = gl_data[(gl_data['AccountType'] == 'Revenue') | (gl_data['AccountType'] == 'COGS')]
salmargin = salmargin.pivot_table(index='Source_id', columns='AccountType', values='AmountNet', aggfunc='sum').fillna(0).round(2)
salmargin['Margin'] = (-salmargin['Revenue'] - salmargin['COGS'])/(-salmargin['Revenue'])
salmargin['Margin%'] = (-salmargin['Revenue'] - salmargin['COGS'])/(-salmargin['Revenue'])*100

salmargin.head(5)
````

Try to understand each column. Is the margin calculated accurately?

````python
salmargin_negative = salmargin[salmargin['Margin%'] < 0]
salmargin_negative.head(5)
````

````python
# We can get a list of all Source_id's with a negative margin using:
salmargin_negative.index
````

````python
# Select orders from gl_data, based on the Source_id being in salmargin_negative
negative_margin_orders = df_revenue[df_revenue['Source_id'].isin(salmargin_negative.index)]
negative_margin_orders = negative_margin_orders[['Booknr', 'JournalDesc', 'Account', 'AmountNet', 'User', 'Period', 'Source_id', 'Relation_name']]
negative_margin_orders.head(10)
````

````python
# We use the pyplot and seaborn libraries to visualize the data, refer to Cheatsheet Week 1a - Libraries and Exploring a dataset
import matplotlib.pyplot as plt
import seaborn as sns

# Analyzing transaction values for outliers
# Define thresholds based on typical transaction ranges
threshold_lower = 20 # Assumed min 20% margin based on knowledge from DigiJazz
threshold_upper = 50 # Assumed max 50% margin based on knowledge from DigiJazz

# Plot histogram for transaction values
plt.figure(figsize=(8, 6))
sns.histplot(salmargin['Margin%'], bins=20)
plt.axvline(threshold_lower, color='red', linestyle='--', label='Lower Threshold')
plt.axvline(threshold_upper, color='blue', linestyle='--', label='Upper Threshold')
plt.title('Distribution of Margins')
plt.xlabel('Margin %')
plt.legend()
plt.show()
````
