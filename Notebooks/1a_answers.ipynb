## 1. Basics: Theory

**Variables**

A _variable_ is a container for storing data values.
- _Naming of a variable_: A variable name can contain only letters, numbers, and underscores. They can start with a letter or an underscore, but not with a number. Spaces cannot be used in variables. The best variable names are informative, such as using ```average_score``` as a variable name for a list of average scores. An example of a bad variable name is ```Average Score```.
- _Assigning values to a variables_: You can assign values (information) to a variable by using the ```=``` sign, for example, you can create the variable ```number_two``` and assign the number ```2``` to this like so: ```number_two = 2```. See below about how you can use the ```print()``` function to access the value assigned to a variable.
-_case-sensitive_: Python is a case-sensitive programming language. This means that it treats uppercase and lowercase letters differently. For example, x and X are not seen as the same variable, student and Student also not the same, etc.
-_overwriting_: Note that if you declare a variable and assign it some value, if you assign the variable again with another value, Python will only use the lastly declared value. In other words, it will overwrite the variable and assign it the new value.

**Data Types**

During this class, we'll work with three data types:
- _String_: A string is a sequence of characters. Strings in Python are surrounded by either single quotation marks, or double quotation marks. Strings can have capital and lowercase letters and spaces in between. An example of a string is ```"Hello World"```.
- _Integer_: An integer is a whole number, positive or negative, without decimals, of unlimited length. An example of an integer is ```0``` or ```1```.
- _Float_: A float, or "floating point number" is a number, positive or negative, containing one or more decimals. An example of a float is ```0.5``` or ```1.0```.

If you want to figure out what data type an object is, you can use the built-in function ```type()```:
- _Type_: The ```type()``` function returns the type of the specified object.

- _Comment_: Comments can be used to explain Python code. They have various purposes, they can be used to make the code more readable and they can be used to prevent execution when testing code. A comment is created by starting a line with a ```#```. An example of a comment is ```# This is a comment.```

**Operators**

It is often convenient to be able to apply operators to your variables. There are many many _operators_ in Python, but you will only need to learn a subset:

- _Addition_: ```x + y```
- _Subtraction_: ```x - y```
- _Multiplication_: ```x * y```
- _Division_: ```x / y```
- _Modulus_: ```x % y```
    - _EXPLANATION_: The modulo operation returns the remainder of a division, after one number is divided by another (called the modulus of the operation). Given two positive numbers x and y, x modulo y (often abbreviated as x mod y) is the remainder of the division of x by y, where x is the dividend and y is the divisor. For example, the expression "5 mod 2" evaluates to 1, because 5 divided by 2 has a quotient of 2 and a remainder of 1, while "9 mod 3" would evaluate to 0, because 9 divided by 3 has a quotient of 3 and a remainder of 0.
- _Exponentiation_: ```x ** y```
    - _EXPLANATION_: Exponentiation is done via the so-called power operator ( ** ) which raises the left value to the power of the second value. For example: 2 ** 3 is 2 times 2 times 2 equals 8.
- _Floor division_: ```x // y```
    - _EXPLANATION_: Floor division is a mathematical operation that is used to divide two numbers and it rounds the result down to the nearest whole number or integer, which is less than or equal to the normal division result. For example, 8/3 has a non-integer value of 2.6667 ; so, applying the floor division operator would return a value of 2 . 
- _Assignment_: ```x = y```
- _Equal_: ```x == y```
- _Not equal_: ```x != y```
- _Greater than_: ```x > y```
- _Less than_: ```x < y```
- _Greater than or equal to_: ```x >= y```
- _Less than or equal to_: ```x <= y```

**Printing results**

Finally, it is useful to be able to print out the result of an operation or the value of a variable on screen so you can inspect if something went wrong. You can do that with the built-in function print.
- _Print_: The ```print()``` function prints the specified message to the screen, or other standard output device. An example of using the ```print()``` function is if we want to access the value we assigned to the ```number_two``` variable above. With ```print(number_two)``` we will access ```2```.

## 2. Datasets & Indexing: Exercises

We will make use of the _Spotify Tracks Dataset_, which contains data about Spotify songs from various different genres. The variables in this dataset are defined below. Read them carefully, as this will be useful during your data analysis.

* _row_number_: The row number of the particular row.
* _track_id_: The Spotify ID for the track.
* _artists_: The artists' names who performed the track.
* _album_name_: The album name in which the track appears.
* _track_name_: Name of the track.
* _popularity_: The popularity of a track is a value between 0 and 100, with 100 being the most popular. The popularity is calculated by algorithm and is based, in the most part, on the total number of plays the track has had and how recent those plays are. Generally speaking, songs that are being played a lot now will have a higher popularity than songs that were played a lot in the past. Duplicate tracks (e.g. the same track from a single and an album) are rated independently. Artist and album popularity is derived mathematically from track popularity.
* _duration_ms_: The track length in milliseconds.
* _explicit_: Whether or not the track has explicit lyrics (```TRUE``` = yes it does; ```FALSE``` = no it does not OR unknown).
* _danceability_: Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable.
* _energy_: Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale.
* _key_: The key the track is in. Integers map to pitches using standard Pitch Class notation. E.g. 0 = C, 1 = C♯/D♭, 2 = D, and so on. If no key was detected, the value is -1.
* _loudness_: The overall loudness of a track in decibels (dB).
* _mode_: Mode indicates the modality (major or minor) of a track, the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0.
* _speechiness_: Speechiness detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks.
* _acousticness_: A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic
instrumentalness: Predicts whether a track contains no vocals. "Ooh" and "aah" sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly "vocal". The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content.
* _liveness_: Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live
valence: A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).
* _tempo_: The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration.
* _time_signature_: An estimated time signature. The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure). The time signature ranges from 3 to 7 indicating time signatures of 3/4, to 7/4.
* _track_genre_: The genre in which the track belongs.

### 3. Compound Data structures: Theory

**Lists**

A Python _list_ is a collection of items, enclosed in ```[ ]``` and separated by commas. Lists can contain different types of data (e.g. strings, integers, float). An example of a list is ```["horse", "cat", "dog"]```.
- _Assigning a list to a variable_: You can assign a list to a variable using a ```=``` (as discussed in Lecture 1), for example, ```animals = ["horse", "cat", "dog"]```.
- _Indexing a list_: Every item in a list is defined by an index, starting from 0. For example, in our ```animals``` list, the item ```"horse"``` has the ```0``` index position and the item ```"cat"``` has the ```1``` index position.
- _Slicing a list_: You can access individual or subset of items from a list through slicing using ```[]``` with the index number(s). For example, you can access the item ```"horse"``` through ```animals[0]```. The sliced subset will always return n-1 of the indexes called. For example, you will return the items ```"horse"``` and ```"cat"``` through ```animals[0:2]```, but if you want all three items you have to call ```animals[0:3]```. You do not have to access items in a sequence, since you can   also use the sytax ```[start:stop:step]```.For example, ```animals[0:3:2]``` will return the items ```"horse"``` and ```"dog"```.
- _Adding items to a list_: You can add items to a list using either ```append()``` or ```insert()```. When using ```append()``` will add the item to the end of the list, for example ```animals.append("tiger")``` results in ```animals = ["horse", "cat", "dog", "tiger"]```. When using ```insert()``` you can specify the location of where you want to add the item with the index, for example, ```animals.insert(1, "tiger")``` results in ```animals = ["horse", "tiger", "cat", "dog"]```. You can also use the ```+``` to add two or more lists together.
- _Deleting items from a list_: You can remove items from a list using ```remove()``` and specifying the list item, for example, after runing ```animals.remove("horse")``` results in just ```["cat", "dog"]```. You can also remove items using the item index and the ```del()``` function, for example, ```del animals[0]``` also results in just ```["cat", "dog"]```.
- _Getting the length of a list_: You figure out how many items are in a list using the ```len()``` function, for example, ```len(animals)``` returns ```3``` indicating that there are three items in the list.

**Tuples**

_Tuples_ are used to store multiple items in a single variable, written with round brackets, ```()```, for example, ```fruits = ("apple", "banana", "cherry", "apple", "cherry")```. Tuples can also be created from lists using ```tuple(list)```. Like lists, tuples can contain multiple data types. Unlike lists, they are collections of elements that are ordered and unchangeable. When we say that tuples are ordered, it means that the items have a defined order, and that order will not change. Tuples are unchangeable, meaning that we cannot change, add or remove items after the tuple has been created. The advantage of using a tuples over lists is that they are faster than lists. Tuples make the code safe from any accidental modification. If a data is needed in a program which is not supposed to be changed, then it is better to put it in tuples than in lists.
- _Indexing a tuple_: Tuples are indexed in the same way as lists.
- _Slicing a tuple_: Tuples are sliced in the same way as lists.
- _Adding items to a tuple_: You can add items to a tuple using ```+```, for example, ```more_fruits = fruits + ("mango")``` results in ```("apple", "banana", "cherry", "apple", "cherry", "mango")```.
- _Deleting items in a tuple_: You cannot remove items from a tuple.
- _Getting the length of a tuple_: You can get the length of a tuple in the same way as for lists.

**Sets**

_Sets_ are used to store multiple elements in a single variable, using ```{}```, for example, ```cars = {"BMW", "Mercedes", "Fiat"}```. A set is a collection which is unordered, unchangeable, and unindexed. A set cannot contain duplicates. In this way, they are useful in identifying differences in two tuples. We will not work with sets in this course.

**Dictionaries**

_Dictionaries_ are used to store data values in ```key:value``` pairs using ```{}```, for example, ```car_info = {"brand": "Ford", "model": "Mustang", "year": 1964}```. A dictionary can also be created by zipping two lists together, using ```dict(zip(list_1, list_2))```. A dictionary is a collection which is ordered, changeable and do not allow duplicates.
- _Indexing a dictionary_: Dictionaries are made up of keys and values, which are indexed either according to their names (e.g. ```"brand"```) or index number (e.g. ```0```). See _Slicing a dictionary_ for more information.
- _Slicing a dictionary_: Dictionaries can be indexed as individual lists by using ```keys()``` and ```values()```. Items can be called individually by first calling either the keys or values as a list using ```list()```. Keys can be called using ```keys()``` and indexed like a list, for example, ```list(car_info.keys())[0]``` will return ```"brand"```. Similarly, values can be called using ```values()``` and indexed like a list, for example, ```list(car_info.values())[0]``` will return ```"Ford"```. Additionally, you can call the value associated with a specific key using ```get()```, for example, ```car_info.get("model")``` returns ```"Mustang"```.
- _Adding items to a dictionary_: You can add a new key:value pair to a dictionary through ```dict[key] = value```, for example, ```car_info["colour"] = "red"``` results in the following dictionary ```car_info = {"brand": "Ford", "model": "Mustang", "year": 1964, "colour": "red"}```.
- _Deleting items from a dictionary_: You can remove an item from a dictionary in the same way as you would delete an item from a list.
- _Changing items in a dictionary_: You can update a dictionary in the same way that you would add items to the dictionary, for example ```car_info["model"] = "Focus"``` results in the following dictionary ```car_info = {"brand": "Ford", "model": "Focus", "year": 1964}```.  
- _Getting the length of a dictionary_: You can get the length of a tuple in the same way as for lists.
