### 2.4. Conditional indexing

You can use operators to specify values that are in line with certain conditions. More specifically, you can type something along the lines of ```df.loc[:,'x']>1``` to get a ```True``` for all the rows of column ```x``` that are higher than 1 and a ```False``` for all the others.  

You can use those Boolean values to show these rows by typing ```df.loc[df.loc[:,'x']>1,'x']```. You could also do this in two steps by first creating an object that, for example, is called _row_indices_, and then using that object. So,
- ```row_indices = df.loc[:,'x']>1```
- ```df.loc[row_indices,'x']```



Q2.4.1. Select the rows that have a Debit amount

````python
# A2.4.1.
print("Method 1 (using loc):")
gl_data.loc[gl_data.loc[:,"Debit/Credit"] == "Debit","Amount"]
````

````python
# A2.4.1.
print("Method 2 (using loc, with an extra step for readability):")
rows_of_interest = gl_data.loc[:,"Debit/Credit"] == "Debit"
gl_data.loc[rows_of_interest,"Amount"]
````

You can create _subsets_ of data in a similar way as shown above, but by _not_ specifying a column. In this case, the ```loc[]``` function is no longer needed. For example, to get _all_ rows of DataFrame ```df``` where, in in column ```x``` the value is greater than 0.5, you can use ```df[df["x"] > 0.5]```.

Note that in this example, you are only subsetting based on rows. You are including _all_ columns.

The benefit of this method is that it reads and writes more easily.

Q2.4.2. Write code in the new way shown, that will provide the same output as the answer for Q10, however, this time showing all columns.

````python
# A2.4.2.
gl_data[gl_data["Debit/Credit"] == "Debit"]
````

You can use this way of conditional indexing to create subsets based on multiple conditions. For example, ```df[(df["x"] > 0.5) | (df["y"] > 0.5)]```, will provide a subset of the ```df``` DataFrame with only the rows where in columns ```x``` **or** ```y``` the values are greater than 0.5. If you want to access data with multiple _and_ conditions, you can use the ```&``` operator.


df[<span style="color:blue">(df["x"] > 0.5)</span> <span style="color:orange">|</span> <span style="color:blue">(df["y"] > 0.5)</span>]
* In <span style="color:blue">Blue, you can see the conditions</span> that are used to subset the data. Notice they are in brackets? This is important to separate the conditions.
* The <span style="color:orange;">|</span> is the _or_ operator. This means that if either of the conditions is met, the row will be included in the subset.


Short recap:
- ```|``` is the _or_ operator
- ```&``` is the _and_ operator
- Brackets `(` and `)` are important to separate the conditions. They work the same way as in mathematics.

### 3.3. Tuples and sets

````python
# We're going to analyze revenue data a few times.

# For convenience, let's create a subset of the data and store this dataframe in a new variable.
# This way, we can quickly access the data we need for the analysis, without having to filter the data every time.

df_revenue = gl_data[gl_data['AccountDesc'] == 'Revenue digital music'].reset_index(drop=True)
df_revenue.head(3)
````

In the dataset, you can find the column ```User```. This column contains the username that posted the transaction. It might be interesting, for our risk analysis, to know how many unique users are in the dataset.

````python
# Create a list of users, and report the first 5
users = list(df_revenue.User)
users[0:5]
````

Q3.3.1. What happens if we convert the ```users``` list into a tuple called ```tuple_users```?

A3.3.1. Using the ```tuple()``` function we can convert the original ```users``` list to a tuple. As you can see, the resulting tuple includes a long list of usernames. It's hard to see how many unique users there are in the dataset.  

````python
#A3.3.1
tuple_users = tuple(users)
print(tuple_users)
tuple_users[0:5]
````

Q3.3.2. What happens if we convert the ```users``` list into a set called ```set_users```? Why is this different from creating a tuple?

A3.3.2) In the set, ```set_users```, only one instance of the each item remains. This is because a set cannot contain duplicates.

````python
#A3.3.2
set_users = set(users)
print(set_users)
````

As you can see, the `set` is a very useful data structure to quickly identify unique values in a dataset. It's important that you understand the difference between a `list`/`tuple` and a `set`, and when to use each of them.
