### 2.4. Conditional indexing

You can use operators to specify values that are in line with certain conditions. More specifically, you can type something along the lines of ```df.loc[:,'x']>1``` to get a ```True``` for all the rows of column ```x``` that are higher than 1 and a ```False``` for all the others.  

You can use those Boolean values to show these rows by typing ```df.loc[df.loc[:,'x']>1,'x']```. You could also do this in two steps by first creating an object that, for example, is called _row_indices_, and then using that object. So,
- ```row_indices = df.loc[:,'x']>1```
- ```df.loc[row_indices,'x']```

You can create _subsets_ of data in a similar way as shown above, but by _not_ specifying a column. In this case, the ```loc[]``` function is no longer needed. For example, to get _all_ rows of DataFrame ```df``` where, in in column ```x``` the value is greater than 0.5, you can use ```df[df["x"] > 0.5]```.

Note that in this example, you are only subsetting based on rows. You are including _all_ columns.

The benefit of this method is that it reads and writes more easily. For that reason, it is often used in the audit practice.
