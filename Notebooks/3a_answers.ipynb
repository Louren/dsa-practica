## Part1: Assessing a multivariate linear regression model

Q1a) Import the ```pandas``` library as pd, import the library ```os```, import train_test_split from sklearn.model_selection.   <br>
Q1b) Read in the dataset ```Staten_Island_housing_market_case.xlsx``` as staten_island_df.<br>
Q1c) Drop the rows of staten_island_df where at least one element is missing.

````python
# A1a)
import pandas as pd
import os
from sklearn.model_selection import train_test_split

#A1b)
staten_island_df = pd.read_excel(os.path.join(os.getcwd(), "Staten_Island_housing_market_case.xlsx"))

#A1c)
staten_island_df = staten_island_df.dropna()
````

Q2a) Create a list with the independent variables called X containing the columns land_sqft ,tot_sqft, yr_built, sale_date, long, lat, atm, bank, and bar.<br>
Q2b) Create a list with dependent variables called y containing only the column price. 

````python
#A2a)
X_features_list = ['land_sqft' , 'tot_sqft', 'yr_built' , 'sale_date' , 'long' , 'lat' , 'atm' , 'bank' , 'bar']
X = staten_island_df[X_features_list]  

#A2b)
y = staten_island_df['price'] 
````

Q3) Create a train and testset for both X and y, names them X_train, X_test, y_train, and y_test. Make sure that the size of the testset is 20% of the total dataset, and use the random state of 1.

````python
#A3)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
````

Q4a) Import mean_squared_error mean_absolute_error from the sklearn.metrics library.<br>
Q4b) Add a constant to the X_train and X_test which is used to add an intercept to your linear model.<br>
Q4c) Run a multivariate linear regression model that fits the model on the trainset using OLS from the sm library.<br>
Q4d) Predict new values of y based on the trainset and for the testset. <br>
Q4e) Calculate MSE, MAE, RMSE, and MAPE for the trainset and for the testset.<br>
Q4f) To what extent do you see any difference in performance between the train and test set? What does that tell you?

````python
#A4a)
from sklearn.metrics import mean_squared_error, mean_absolute_error
import statsmodels.api as sm

#A4b)
X_test = sm.add_constant(X_test)
X_train = sm.add_constant(X_train)

#A4c)
multivar_linreg = sm.OLS(y_train, X_train).fit()

#A4d)
y_pred_test = multivar_linreg.predict(X_test)
y_pred_train = multivar_linreg.predict(X_train)

#A4e)
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)

mae_train = mean_absolute_error(y_train, y_pred_train)
mae_test = mean_absolute_error(y_test, y_pred_test)

rmse_train = mse_train**0.5
rmse_test = mse_test**0.5

mape_train = ((y_train - y_pred_train) / y_train).abs().mean() * 100
mape_test = ((y_test - y_pred_test) / y_test).abs().mean() * 100

#A4f)
print("Mean Squared Error train:", mse_train, "test:", mse_test)
print("Mean Absolute Error train:", mae_train, "test:", mae_test)
print("Root Mean Squared Error train:", rmse_train, "test", rmse_test)
print("Mean Absolute Percentage Error train:", mape_train, "test:", mape_test)
````

A4f) The MAE, MSE, RMSE and MAPE of the regression model are very similar on the train and test set. This means that the model generalizes well to unseen data (test set). In modelling terms that means that there is no "overfitting".

Q4g) Play around with adding more variables or removing them, what happens to the model performance on train and testset?

A4g) This depends on the variables that you add or remove, but in general adding variables to your model improves performance on the trainset. It can also improve performance on the testset, but that depends on whether you're already overfitting or not.

## Part2: Assessing a KNN model

Q5a) Import KNeighborsClassifier from the library sklearn.neighbors, import MinMaxScaler from the library sklearn.preprocessing, and import classification_report and confusion_matrix from sklearn.metrics <br>
Q5b) Like in the previous lecture, we'll be predicting the type of home. To make sure there are only two classes, remove all three family homes from the dataset. You can find them in the variable bldg_ctgy.<br>
Q5c) Create your target variable y containing bldg_ctgy, and your feature variable X containing the features: land_sqft, tot_sqft, yr_built, sale_date, long, lat, atm, bank, bar.<br>
Q5d) Create a train and testset for both X and y, names them X_train, X_test, y_train, and y_test. Make sure that the size of the testset is 20% of the total dataset, and use the random state of 1.<br>
Q5e) Normalize the feature data by using the MinMaxScaler, and create a scaled version of your X_train and X_test (name the X_train_scaled, and X_test_scaled).<br>
Q5f) Train your knn model on the scaled data, use 4 neighbors to predict class.<br>
Q5g) Compute the predictions of the model based on the traindataset, and show the confusion matrix. <br>
Q5h) Compute the predictions of the model based on the testset, and show the confusion matrix.<br>
Q5i) What do you think? Does this model perform well on the trainset?

````python
#Q5a)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report, confusion_matrix

#Q5b)
staten_island_df = staten_island_df[staten_island_df.bldg_ctgy != '03  THREE FAMILY HOMES']

#Q5c)
y = staten_island_df['bldg_ctgy']
X_features_list = ['land_sqft', 'tot_sqft', 'yr_built', 'sale_date', 'long', 'lat', 'atm', 'bank', 'bar']
X = staten_island_df[X_features_list]

#Q5d) 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 1)

#Q5e)
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#Q5f)
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train_scaled, y_train)

#Q5g)
y_pred_train = knn.predict(X_train_scaled)
conf_matrix = confusion_matrix(y_train, y_pred_train)
unique_classes = sorted(set(y_train))
conf_matrix_df = pd.DataFrame(conf_matrix, index=unique_classes, columns=unique_classes)
conf_matrix_df.index.name = 'Actual'
conf_matrix_df.columns.name = 'Predicted'
print(conf_matrix_df)

#Q5g)
y_pred_test = knn.predict(X_test_scaled)
conf_matrix = confusion_matrix(y_test, y_pred_test)
unique_classes = sorted(set(y_test))
conf_matrix_df = pd.DataFrame(conf_matrix, index=unique_classes, columns=unique_classes)
conf_matrix_df.index.name = 'Actual'
conf_matrix_df.columns.name = 'Predicted'
print(conf_matrix_df)
````

Q6a)  We can also calculate the accuracy, precision and recall using functions. Calculate the accuracy, precision and recall for the trainset. Import accuracy_score, precision_score, recall_score from the library sklearn.metrics.<br>
Q6b) In the code you see how to apply the functions for accuracy, precision, and recall. How do you think the model performs after you ran the functions?<br>
Q6c) Apply the same functions, but now on the testset with the predicted target variable. What do you think of the model now?<br>

````python
#Q6a)
from sklearn.metrics import accuracy_score, precision_score, recall_score

#Q6b)
accuracy_train = accuracy_score(y_train, y_pred_train)
precision_train = precision_score(y_train, y_pred_train, average='binary', pos_label='01  ONE FAMILY HOMES')
recall_train = recall_score(y_train, y_pred_train, average='binary', pos_label='01  ONE FAMILY HOMES')
print("Train accuracy: ", accuracy_train)
print("Train precision: ", precision_train)
print("Train recall: ", recall_train)

# Q6c)
accuracy_test= accuracy_score(y_test, y_pred_test)
precision_test = precision_score(y_test, y_pred_test, average='binary', pos_label='01  ONE FAMILY HOMES')
recall_test = recall_score(y_test, y_pred_test, average='binary', pos_label='01  ONE FAMILY HOMES')
print("Test accuracy: ", accuracy_test)
print("Test precision: ", precision_test)
print("Test recall: ", recall_test)
````

Q6c) It is not bad, the measures are a bit lower as expected, but performance drops only a bit. Hence, the model generalizes quite well to the test set.
